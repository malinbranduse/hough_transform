# Image Preprocessing

In this stage the aim is to convert the input image to binary edges using the Canny algorithm

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

After the used libraries are imported, the next step is to read the image based on the provided path and name:

```python
input_dir = './input/'
in_img_name = 'Kheops-Pyramid.jpeg'
in_img = cv.imread(input_dir + in_img_name, 0)
```

Displaying the read image as gray scale:

```python
plt.figure(figsize=(10,10))
plt.imshow(in_img,cmap = 'gray')
plt.title('The gray scale input image')
plt.show()
plt.figure(figsize=(10,5))
plt.hist(in_img.ravel(),256,[0,255])
plt.suptitle('The brightness histogram')
plt.show()
```

Using the canny function from opencv, the binary image is extracted. The canny hysteresis threshold have been chosed experimentally for getting the best edge selection in the image

```python
canny_thresholds = [200, 220]
edges = cv.Canny(in_img, canny_thresholds[0], canny_thresholds[1])
plt.figure(figsize=(15,10))
plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image')
plt.show()
```

Displaying the histogram, the binary image can be observed as having just black and white pixels:

```python
plt.hist(edges.ravel(),256,[0,255])
plt.suptitle('The brightness histogram')
plt.show()
```
